```python
#API Information
import tweepy
from tweepy import OAuthHandler

consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''
 
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
 
api = tweepy.API(auth)
```

```python
twitterHandles = ['drfeifei','demishassabis','goodfellow_ian','iamtrask','sirajraval','karpathy','AndrewYNg','ylecun',
                  'fchollet','OpenAI','elonmusk','BaiduResearch','googleresearch','pabbeel','DeepMindAI','romanyam',
                  'hugo_larochelle']
print(len(twitterHandles))
```

```python
#quick check to ensure no spelling errors
for handle in twitterHandles:
    print(api.get_user(handle).screen_name)
```

```python
import random
import time 
import csv
from __future__ import print_function


fpath = 'tweets.txt'
alltweets = []
maxTweets = 200

for handle in twitterHandles:
    
    print("\nreading tweets from: @" + handle + "\n")
    
    new_tweets = api.user_timeline(screen_name=handle,count=maxTweets)
    alltweets.extend(new_tweets) #save recent tweets
    oldest = alltweets[-1].id - 1 #save oldest tweets minus 1

    while len(new_tweets) > 0:
        print("getting tweets before: " + str(oldest))
        
        new_tweets = api.user_timeline(screen_name = handle,count=200,max_id=oldest)
        alltweets.extend(new_tweets)
        oldest = alltweets[-1].id -1

        print(str(len(alltweets))+ " tweets downloaded so far")
    
    
#write tweets
with open(fpath,'wb') as f:
    for tweet in alltweets:
        print(tweet.text.encode("utf-8"),file=f)       
```
